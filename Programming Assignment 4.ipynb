## ECE Board Problem. Using data wrangling and data visualization technique  analyze the data and presen different  dataframes and  visuals using the dataset given.

### 1a data wrangling. creating a dataframe based on the given format Vis = [“Name”, “Gender”, “Track”, “Math<70”]; hometown is constant as Visayas. Content of the dataframe to be made should be Name, Grades for GEAS, and Grades for Electronics that are greater than 70.

```python
#ECE Board Problem 1a
#setting Instru as the dataframe which only includes the conditions that the track is Instrumentation, Hometown is Luzon, and Grade in electronics is greater than 70
Instru = df[(df['Track']=='Instrumentation') & (df['Hometown']=='Luzon') & (df['Electronics']>70)]

#setting instru to print and display the needed columns as per the given format
instru= Instru[['Name', 'GEAS', 'Electronics', 'Hometown']]

#printing and displaying the new dataframe
instru
```

### 1b data wrangling. creating a dataframe based on the given format Vis = [“Name”, “Gender”, “Track”, “Math<70”]; hometown is constant as Visayas. Content of the dataframe to be made should be Name, Track, and Grades for Electronics, and Average Grade for all subjects.

```python
#ECE Board Problem 1b
#getting the average of all of the subjects in the given dataset
df['average'] = ((df['Math'])+(df['Electronics'])+(df['GEAS'])+(df['Communication']))/4

#setting mindy as a new dataframe which only consists of the data with conditions given which is Mindanao as hometown, Female as gender, and the average grade is greater than or equal to 55.
Mindy = df[(df['Hometown']=='Mindanao') & (df['Gender']=='Female') & (df['average']>=55)]

#setting Mindy to be the columns needed following the given format
Mindy = Mindy[['Name', 'Track', 'Electronics', 'average']]

#printing and displaying the new dataframe
Mindy
```

### 2 data visualization. The goal is to create a data visualization that includes the following; chosen track in college, gender, or hometown and see if any of them contributes to getting a higher score. A box plot was chosen here to make the visualization of the outliers easier. The matplotlib and seaborn libraries were used for the visualisation.

### Importing the matplotlib and seaborn libraries 

```python
#Problem 2
import matplotlib.pyplot as plt
import seaborn as sns
```

### The code for the box plot visualization using the 2 libraries mentioned

```python
#setting the size of the figure
plt.figure(figsize=(8, 6))

#setting which dataframe the boxplot would use and which data would be displayed on the axis as well as what the colors represesnt
sns.boxplot(x='Track', y='average', hue='Gender',data=df)

#labeling the y axis as Average which contains the average scores of the students
plt.ylabel('Average')

#labeling the x axis as Chosen track to see which track the students are part of
plt.xlabel('Chosen Track')

#making the title for the whole graph 
plt.title('Average Grades of students across different tracks and Gender')
```
